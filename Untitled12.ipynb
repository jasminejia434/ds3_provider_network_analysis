```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/content/tin_npi_1.csv'
df = pd.read_csv(file_path)

df['Geographical_Location'] = df['PROVIDER_BUSINESS_MAILING_ADDRESS_STATE_NAME']
df['Specialty'] = df['PROVIDER_TAXONOMY_DESCRIPTION']

# Count the number of NPIs by top states/Specialty
top_states = df['Geographical_Location'].value_counts().head(10)
top_specialties = df['Specialty'].value_counts().head(10)

# Plotting top states
plt.figure(figsize=(10, 6))
sns.barplot(x=top_states.values, y=top_states.index, hue=top_states.index, dodge=False, legend=False)
plt.xlabel('Number of NPIs')
plt.ylabel('State')
plt.title('Top 10 States by Number of NPIs')
plt.tight_layout()
plt.show()

# Plotting top states
plt.figure(figsize=(10, 4))
sns.barplot(x=top_specialties.values, y=top_specialties.index, hue=top_specialties.index, dodge=False, legend=False)
plt.xlabel('Number of NPIs')
plt.ylabel('Specialty')
plt.title('Top 10 Specialties by Number of NPIs')
plt.tight_layout()
plt.show()

```

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Take a sample of the data to reduce computation time
sample_size = 3000
df = df.sample(n=sample_size, random_state=42)

# Extract NPI and TIN_VALUE columns
npi_list = df['NPI'].astype(str).tolist()
tin_list = df['TIN_VALUE'].astype(str).tolist()

# Create a bipartite graph
B = nx.Graph()

# Add nodes with the bipartite attribute
B.add_nodes_from(npi_list, bipartite='NPI')
B.add_nodes_from(tin_list, bipartite='TIN')

# Add edges between NPIs and TINs
edges = list(zip(npi_list, tin_list))
B.add_edges_from(edges)

print(B)

# Separate nodes by set
npi_nodes = {n for n, d in B.nodes(data=True) if d['bipartite'] == 'NPI'}
tin_nodes = set(B) - npi_nodes

# Clustering analysis for NPIs and TINs
# Create a dataframe for clustering with NPI and TIN relationships
adjacency_matrix = nx.to_pandas_adjacency(B, nodelist=list(npi_nodes) + list(tin_nodes))

# Limit the size of the adjacency matrix to avoid issues with large graph size
adjacency_matrix = adjacency_matrix.loc[:, (adjacency_matrix != 0).any(axis=0)]

# Standardize the data
scaler = StandardScaler()
adjacency_matrix_scaled = scaler.fit_transform(adjacency_matrix)

# Elbow plot to determine optimal number of clusters
sse = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(adjacency_matrix_scaled)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_range, sse, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Plot to Determine Optimal Number of Clusters')
plt.tight_layout()
plt.show()
```

```python
# Visualize the network graph with different colors for bipartite sets
plt.figure(figsize=(15, 15))
pos = nx.spring_layout(B, seed=42)
npi_color = 'blue'
tin_color = 'red'
nx.draw_networkx_nodes(B, pos, nodelist=npi_nodes, node_color=npi_color, node_size=10, label='NPI')
nx.draw_networkx_nodes(B, pos, nodelist=tin_nodes, node_color=tin_color, node_size=10, label='TIN')
nx.draw_networkx_edges(B, pos, edge_color='gray')
plt.title('Network Graph of NPIs and TINs')
plt.legend()
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Network properties
# 1) Degree distribution for NPIs and TINs
degrees_npi = [B.degree(n) for n in npi_nodes]
degrees_tin = [B.degree(n) for n in tin_nodes]

plt.hist(degrees_npi, bins=20, alpha=0.5, label='NPI')
plt.hist(degrees_tin, bins=20, alpha=0.5, label='TIN')
plt.xlabel('Degree')
plt.ylabel('Nodes')
plt.title('Degree Distribution for Bipartite Network')
plt.legend()
plt.show()

# 2) Density
print("Density of bipartite graph:", nx.density(B))

# 3) Connected components
connected_components = list(nx.connected_components(B))
print("Number of connected components:", len(connected_components))

# Node properties
# 1) Clustering coefficient
# Note: Clustering coefficient is typically not defined for bipartite graphs, but we can calculate it for the projection
npi_projection = nx.projected_graph(B, npi_nodes)
print("Average clustering coefficient for NPI projection:", nx.average_clustering(npi_projection))

# 2) Centrality
betweenness = nx.betweenness_centrality(B)
betweenness_df = pd.DataFrame.from_dict(betweenness, orient='index', columns=['Centrality'])
print(betweenness_df.head())
```

```python
from collections import defaultdict

# Find clusters: Identify NPIs that are connected to the same TIN
npi_groups_by_tin = defaultdict(list)
for tin in tin_nodes:
    connected_npis = list(B.neighbors(tin))
    if len(connected_npis) > 1:
        npi_groups_by_tin[tin] = connected_npis

# Create a dictionary to store NPI attributes
npi_attributes = df.drop_duplicates(subset='NPI').set_index('NPI')[['Geographical_Location', 'Specialty']].to_dict('index')

# Find the top 1 TINs with the highest number of associated NPIs
top_1_tins = sorted(npi_groups_by_tin.items(), key=lambda item: len(item[1]), reverse=True)[:1]

# Print the top 1 TINs and their associated NPIs
print("Top 1 TINs with the highest number of NPIs:")
for tin, npis in top_1_tins:
    print(f'TIN: {tin}, Number of NPIs: {len(npis)}')
    print(f'  NPIs: {npis}')
    print()


```
